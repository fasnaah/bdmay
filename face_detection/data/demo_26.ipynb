```python
import cv2
img=cv2.imread('/home/fasna/Music/bdmay/face_detection/data/WhatsApp Image 2023-12-26 at 8.57.29 AM.jpeg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
face_cascade=cv2.CascadeClassifier('/home/fasna/Music/bdmay/face_detection/data/haarcascade_frontalface_default.xml')
```

```python
faces=face_cascade.detectMultiScale(img_gray)
```

```python
len(faces)
```

```python
faces
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)
```

```python
cv2.imshow('face detection',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
